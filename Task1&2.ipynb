```python
import os
import json
from datetime import datetime
```

## Task 1

```python
x=input('Enter file name: [For example: key_metrics.json]  \n')
if(os.path.isfile(x)):
    f = open(x,'r' )
    # returns JSON object as a dictionary
    data = json.load(f)
    f.close()
    counts = data['report']['data']
    list_metrics_counts=[]
    metrics_counts={}
    metrics_list=[]
    for i in data['report']['metrics']:
        metrics_list.append(i['name'])
    for i in counts:
        metrics_counts={}
        metrics_counts['date'] = datetime(i['year'],i['month'],i['day'])
        for j in range(len(metrics_list)):
            metrics_counts[metrics_list[j]] = i['counts'][j]
        list_metrics_counts.append(metrics_counts)
    
    print(list_metrics_counts)
    
    jsonString = json.dumps(list_metrics_counts,default=str)
    jsonFile = open('task_1.json', "w") #save as json file "task_1"
    jsonFile.write(jsonString)
    jsonFile.close()
else:
    print("Invalid input file name, Please enter correct file name")
```
